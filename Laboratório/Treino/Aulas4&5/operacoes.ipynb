Exercícios de Reposição -- Operações Aritméticas e Lógicas

```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

# Aplicar HDR combinando as quatro imagens:

flor1 = Image.open('Imagem1.jpeg')
flor2 = Image.open('Imagem2.jpeg')
flor3 = Image.open('Imagem3.jpeg')
flor4 = Image.open('Imagem4.jpeg')

flor1_np = np.array(flor1)
flor2_np = np.array(flor2)
flor3_np = np.array(flor3)
flor4_np = np.array(flor4)

# Normalizar as imagens para o intervalo [0, 1]
flor1_norm = flor1_np / 255.0
flor2_norm = flor2_np / 255.0
flor3_norm = flor3_np / 255.0
flor4_norm = flor4_np / 255.0

# Combinar as imagens usando a média
hdr_image = (flor1_norm + flor2_norm + flor3_norm + flor4_norm) / 4

plt.imshow(hdr_image)
plt.title('Imagem HDR')
plt.savefig('hdr_image.png')  # Salvar a imagem HDR resultante

```

Questão 2 - Aumentar e Diminuir o brilho em 40%

```python
# Aumentando o brilho da imagem HDR
hdr_image_bright = np.clip(hdr_image * 1.4 * 255, 0, 255).astype(np.uint8) # Multiplicando por 255 para "desnormalizar" a imagem

plt.subplot(2, 1, 1)
plt.imshow(hdr_image_bright)
plt.title('Imagem HDR com Brilho Aumentado')

# Aumentando espaço entre as imagens
plt.subplots_adjust(hspace=0.5)

# Reduzindo o brilho da imagem HDR
hdr_image_dark = np.clip(hdr_image * 0.6 * 255, 0, 255).astype(np.uint8) # Multiplicando por 255 para "desnormalizar" a imagem
plt.subplot(2, 1, 2)

plt.imshow(hdr_image_dark)
plt.title('Imagem HDR com Brilho Reduzido')
plt.savefig('hdr_image_brightness_switch.png')
```

Questão 3 - NOT na imagem da questão 1

```python
# Aplicando not na imagem da questão 1
hdr_image_not = np.clip(1 - hdr_image, 0, 1) * 255 # Aplicando NOT e multiplicando por 255 para "desnormalizar" a imagem
hdr_image_not = hdr_image_not.astype(np.uint8)
plt.imshow(hdr_image_not)
plt.title('Imagem HDR com NOT Aplicado')
plt.savefig('hdr_image_not.png')
```
