```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = Image.open('luffy.png')
plt.imshow(img)

imgnp = np.array(img)
imgLuffy = imgnp[:,:,:3] # Retirar o canal alpha

# Voltar para imagem:

imgLuffy1 = Image.fromarray(imgLuffy)
plt.imshow(imgLuffy1)

```

```python
# Cores RGB

# Método Average para Grayscale
l, c, p = imgnp.shape

img_avg = np.zeros(shape=(l,c), dtype=np.uint8)

for i in range(l):
    for j in range(c):
        r = float(imgnp[i,j,0])
        g = float(imgnp[i,j,1])
        b = float(imgnp[i,j,2])

        img_avg[i,j] = (r + g + b) / 3


plt.imshow(img_avg, cmap= 'gray')
```

```python
img_avg2 = np.average(imgnp[:,:,:3], axis = 2).astype(np.uint8)
plt.imshow(img_avg2, cmap = 'gray')
```

```python
img_np_float64 = imgnp[:,:,:3].astype(np.float64)
img_avg3 = ((img_np_float64[:,:,0] + img_np_float64[:,:,1] + img_np_float64[:,:,2]) / 3).astype(np.uint8)
plt.imshow(img_avg3, cmap='gray')
```

## Imagem Binária (Preto e Branco)

```python
limiar = 128

for i in range(l):
    for j in range(c):

        r = float(imgnp[i, j, 0])
        g = float(imgnp[i, j, 1])
        b = float(imgnp[i, j, 2])
        
        img_avg[i,j] = (r + g + b) / 3
    
        if img_avg[i,j] > limiar:
            img_avg[i,j] = 255
        else:
            img_avg[i,j] = 0
            

plt.imshow(img_avg, cmap='gray')
```
